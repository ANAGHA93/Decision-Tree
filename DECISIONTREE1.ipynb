```python
df=df.rename(columns={'Marital.Status':'MI','Taxable.Income':'TaxableIncome','Work.Experience':'WorkExperience'})
```

```python
data_final = df.iloc[:,0:6].join(x)
data_final
```

```python
data_final=data_final.drop(['Urban'],axis=1)
```

```python

data_final1 = data_final.iloc[:,2:9].join(y)
data_final1
```

```python
data_final2= data_final1.iloc[:,0:8].join(z)
data_final2
```

```python
#finding the maximum value of TaxableIncome
tax_max=data_final2['TaxableIncome'].max(axis=0)
tax_max
```

```python
#finding the minimum value of TaxableIncome
tax_min=data_final2['TaxableIncome'].min(axis=0)
tax_min
```

```python
#The output variable Taxgroup created and contents are made by splitting the TaxableIncome variable where TaxableIncome between 10003 and30000 are categorised as  "Risky" and between 30000 and 99619 as "Good"
data_final2['Taxgroup']=pd.cut(data_final2.TaxableIncome,bins=[10003,30000,99619],labels=["Risky","Good"])
data_final2
```

```python
#Label encoding is done for the output variable Taxgroup
label_encoder = preprocessing.LabelEncoder()
data_final2['Taxgroup']= label_encoder.fit_transform(data_final2['Taxgroup'])
```

```python
#Input and Output variables are seperated
x=data_final2.iloc[:,0:10]
y=data_final2["Taxgroup"]
```

```python
# importing the CART DecisionTree classifier
from sklearn.tree import DecisionTreeClassifier
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
```

```python
#splitting to testing and training datas
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
#Building the  Model on the test data
model_gini.fit(x_train, y_train)
```

```python
#Predicting on test data
preds = model_gini.predict(x_test) # predicting on test data set 
pd.Series(preds).value_counts()
```

```python
#Prediction and computing the accuracy
pred=model_gini.predict(x_test)
np.mean(preds==y_test)
```
