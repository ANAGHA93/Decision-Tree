```python
#Read the file
df=pd.read_csv('Company_Data.csv')
```

```python
#so if the sales>8 we consider it as high sale
df['SalesHigh']=pd.cut(df.Sales,bins=[0,8,17],labels=["NO","YES"])
df
```

```python
#Dropping the Sales column from the data since we used it to make the ouput variable SalesHigh which shows whether Sales value is High or not 
data=df.drop(['Sales'],axis=1)

```

```python
#Label encoding done on output variable
label_encoder = preprocessing.LabelEncoder()
data3['SalesHigh']= label_encoder.fit_transform(data3['SalesHigh'])
```

```python
datay=datas.rename(columns={'SalesHigh':'SalesHigh1'})
```

```python
DATA=data4.rename(columns={'SalesHigh1':'SalesHigh'})
```

```python
#This is our final data
DATA
```

```python
#Seperating the input and output variables
x=DATA.iloc[:,0:15]
y=DATA["SalesHigh"]
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
#Importing the  CART DecisionTree Classifier
from sklearn.tree import DecisionTreeClassifier
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
```

```python
#Building the model on training data
model_gini.fit(x_train, y_train)
```

```python
#Predicting on test data
preds = model_gini.predict(x_test) # predicting on test data set 
pd.Series(preds).value_counts()
```

```python
pd.crosstab(y_test,preds)
```
